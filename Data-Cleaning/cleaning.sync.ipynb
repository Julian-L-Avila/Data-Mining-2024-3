```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = "./salaries.csv"

df = pd.read_csv('./salaries.csv')

df.info()
```

```python
chunk_size = 200000

dataframes = []
nullcounts = 0.0

for chunk in pd.read_csv(data, chunksize=chunk_size):
    df = pd.DataFrame(chunk)

    print(df.info())
    dataframes.append(df)

    nullcounts += df.isnull().sum()

print(nullcounts)
```
